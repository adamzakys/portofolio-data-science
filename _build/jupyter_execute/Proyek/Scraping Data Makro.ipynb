```python
import csv
from bs4 import BeautifulSoup
import os
import shutil
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')



file_path = os.path.join('/content/drive', 'My Drive', 'Skripsi', 'Bahan', 'KalenderMakroekonomi2.html')
```

```python

with open(file_path, 'r') as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, 'html.parser')

rows = soup.find_all('tr', class_='js-event-item')

data = []
for row in rows:
    date = row['data-event-datetime'].split()[0]
    currency = row.find('td', class_='flagCur').text.strip()
    event = row.find('td', class_='event').text.strip()
    actual_element = row.find('td', class_=lambda c: c and 'actual' in c)
    if actual_element:
        actual = actual_element.text.strip()
    else:
        actual = "Not Found"

    data.append([date, currency, event, actual])  # Append only desired columns

with open('data_ekstrak.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Tanggal', 'Mata Uang', 'Event', 'Aktual'])  # Updated header
    writer.writerows(data)

print("Data telah disimpan ke file 'data_ekstrak.csv'")
```

```python
data_frame = pd.DataFrame(data, columns=['Tanggal', 'Mata Uang', 'Event', 'Aktual'])
data_frame.to_csv("DF-Makroekonomi2.csv",index=False)
```

```python
source_file = '/content/DF-Makroekonomi2.csv'
destination_folder = '/content/drive/My Drive/Skripsi/Bahan/'

os.makedirs(destination_folder, exist_ok=True)

destination_file = os.path.join(destination_folder, 'DF-Makroekonomi2.csv')

shutil.copy2(source_file, destination_file)

print(f"File '{source_file}' copied to '{destination_file}'")
```
